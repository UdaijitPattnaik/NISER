```python
'''
Ask the user for a number. 
Find all the even and odd numbers in between
1 to "number". 

Define a function that separates the numbers into 2 lists and returns the 2 lists. 

Print individual lists returned from the function. 
'''


def oddevenlist(tillrange):
  # create an odd list
  oddlist = []
  # create an even list
  evenlist = []

  # populate the lists with even and odd numbers
  # loop through all the numbers
  for number in range(1,tillrange): 
    # if number is even or if it is divisble by 2  
    if number%2 == 0:
      # then put into even list
      evenlist.append(number)
    # if number is odd or else
    else:
      # then put into odd list
      oddlist.append(number)
  return oddlist, evenlist

tillrange = 11
ioddlist, ievenlist = oddevenlist(tillrange)
print(ioddlist)
print(ievenlist)
```

```python
# Creating a simple class

class Student:
  
  def __init__(self, name, age):
    # Special method, that is invoked whenever a new instance is created. 
    """
    The self parameter is required in the method definition, and it must 
    come first before the other parameters. It must be included in the 
    definition because when Python calls this __init__() method later 
    (to create an instance of Dog), the method call will automatically 
    pass the self argument. Every method call associated with a class 
    automatically passes self, which is a reference to the instance itself; 
    it gives the individual instance access to the attributes and methods 
    in the class.
    # Eric Matthes - Python Crash Course
    """
    self.name = name
    self.age = age
  
  
  def printInfo(self):
    print("Student name is ", self.name, "and age is ", self.age)
  
  def yearstohundred(self):
    print(self.name, "has ", 100-self.age, "years to become 100.")
  
```
