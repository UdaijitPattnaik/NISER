```python
def binarysearch(nums,low,high, key):
    if low <= high:
        mid = (low + high)/2
        mid = int(mid)
        if nums[mid] == key:
            return mid
        elif nums[mid] > key:
            return binarysearch(nums,low,mid-1,key)
        else:
            return binarysearch(nums,mid+1,high,key)
        
    else:
        return -1



numlist = [5,6,10,54,89,897,587,568,214]
print('unsorted list is:',numlist)
numlist.sort()
print('sorted list :',numlist)

```

```python

n=int(input('enter key to search: '))
key = n
out = binarysearch(numlist,0,len(numlist)-1,key)
if out > -1:
  print(key, 'its index is', out)
else:
  print(key, 'is not in the list')
```
