```python
def prime(n):
  c=0
  for i in range(2,n//2):
    if n%i == 0:
      c=c+1
  if c>0:
    return False
  else:
    return True
limit= int(input('enter the limit:'))
v=0
for j in range(1,limit+1):
  a= prime(j)
  if a is True:
    v=v+1
    

print('the number of primes in range 1 to',limit,'is',v)

```
