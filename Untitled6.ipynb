```python
n=int(input('enter the value of n'))
sumx=0
sumy=0
sumxy=0
sumxs=0
x=0
y=0
for i in range(1,n+1):
  x= float(input('enter the value of x'))
  y= float(input('enter the value of y'))
  xs= x*x
  print('xsquare=',xs)
  xy= x*y
  print('xy=',xy)
  sumx=sumx+x
  sumy=sumy+y
  sumxy=sumxy+xy
  sumxs=sumxs+xs
print('sumx=',sumxs,'sumy=',sumy,'sumxy=',sumxy,'sumxs=',sumxs)
d=(n*sumxs)-((sumx)*(sumx))
print('d=',d)
m=((n*sumxy)-sumxy)/d
print('m=',m)
c=((sumxs*sumy)-(sumx*sumxy))/d
print('c=',c)
```
