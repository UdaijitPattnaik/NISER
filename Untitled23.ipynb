```python
import torch
from torch import Tensor
import numpy as np
import matplotlib.pyplot as plt

# Set the number of simulations to run
num_simulations = 100

# Set the number of localized states and dopants per simulation
num_localized_states = 10000000
num_dopants = 1

# Set the concentration of ionized dopant atoms
Nd = 1

# Initialize an array to store the distances from all simulations
all_distances = torch.empty(num_simulations * num_localized_states)

# Run multiple simulations
for i in range(num_simulations):
    # Generate random positions for localized states and dopants
    localized_states = torch.normal(mean=0.5, std=0.2, size=(num_localized_states, 2))
    dopants = torch.rand(size=(num_dopants, 2))

    # Find the distance to the nearest dopant for each localized state
    distances = torch.cdist(localized_states, dopants).min(dim=1).values

    # Store the distances from this simulation
    all_distances[i*num_localized_states:(i+1)*num_localized_states] = distances

# Plot the average empirical probability density of finding the nearest dopant at any distance r
plt.hist(all_distances.numpy(), bins=200, density=True, label='Empirical')

# Define the theoretical probability density function
def wsrd(r: Tensor) -> Tensor:
    return 4 * np.pi * r**2 * Nd * np.exp(-4 * np.pi * r**3 * Nd / 3)

# Plot the theoretical probability density function
r = torch.linspace(0, 1.5, 100)
plt.plot(r.numpy(), wsrd(r).numpy(), label='Theoretical')

# Add labels and a legend
plt.xlabel('Distance to nearest dopant')
plt.ylabel('Probability density')
plt.legend()

# Show the plot
plt.show()
```
