```python
import numpy as np
A= np.array([[0,1,1,-2],[1,2,-1,0],[2,4,1,-3],[1,-4,-7,-1]])
B= np.array([[-3],[2],[-2],[-19]])
l=len(A)
print(A);print(B)
for i in range(0,l):
    if A[i][i]==0:
        max =i
        for j in range(i+1,l):
            if A[j][i]>A[max][i]:
                max = j
        SA=np.array([[0,0,0,0]])
        SB=np.array([[0,0,0,0]])
        SA[0]=A[max]
        s=B[max][0]
        A[max]=A[i]
        B[max][0]=B[i][0]
        A[i]=SA[0]
        B[i][0]=s
    if A[i][i] != 1:
        A[i]=A[i]/A[i][i]
        B[i]=B[i]/A[i][i]
        print(A)
        print(B)
    for j in range(i+1,l):
        if A[j][i] != 0:
            A[j]=A[j]-A[j]*A[i]
            B[j]=B[j]-B[j]*B[i]
print(A); print(B)
```
